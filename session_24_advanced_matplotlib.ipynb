```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# we can change the color of code i means alag alag color provide kr skte hai
# first step hame  ispe graph plot kr rha he unko ek number assigned krna padega
# iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})->aise ham assign kr skte hai number
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
# c=iris['Species']->using this ham color code kr skte hai
# cmap='color_code name in matplotlib'[->using this we can change color]
# plt.colorbar()->will tell ki kaun sa color kya represent kr rha hai
# alpha=num->this will decide ki mere graph ka opacity kya haii mean translucent hai ya nhi
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# cmap and alpha
```

### Plot size

```python
# plt.figure(figsize=(lenght,width))->using this  ham grpah ki size inc or dec kr skte hai
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Subplots

```python
# A diff way to plot graphs
# using subplot ham ek sath more than one graph plot kr skte hai
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# when we write plt.subplot() then fig->jaha graph plot honge  and axe->isme wo hoga jis grpag plot hoga  call ho jata hai
fig,ax = plt.subplots(figsize=(15,6))
# niche wale se scatter graph plot kr skte hai
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
# using this ham tittle de skte graph ka
ax.set_title('Something')
#  using this ham x axis ka name de skte hai
ax.set_xlabel('Avg')
# using this ham y axis ka name de skte hai
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# batter dataset
```

```python
# nrows=2,ncols=1 ->this will decide ki kitne graph banenge ek sath
# sharex= true->it is parameter jo x ka name dono graph ke share hoga taki massey na ho
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
# here this is like vector so [0,0] is most left uppar wala and so on
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
# using this we can also draw subplot (2,2,1)->(norow,nocol,nofgraph)
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
# meshgrid(x,y)->ye ek grid form kr dega
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
# colorbar(p)->using this jaha value jyada waha ka color will differ from other
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots

```python
# using contour plots ham 3d ko 2d mein represent krte hai
# contour(xx,yy,z,cmap='viridis')->just we only wrote contour and this will be plotted
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
# imshow(grid)->usiing this we can plot heatmap using grid data
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
